```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

from sklearn.ensemble import RandomForestRegressor 
from sklearn.model_selection import cross_val_score

# Load 
df = pd.read_csv("/kaggle/input/datasets/karansridhar/marketing-and-product-performance/marketing_and_product_performance.csv")

X = df.select_dtypes(include=np.number).drop(columns=['Revenue_Generated']) 
y = df['Revenue_Generated'] 

model = RandomForestRegressor()

scores = cross_val_score(model, X, y, cv=5, scoring='r2')

table = pd.DataFrame({ 
    "Fold": range(1,6), 
    "R2": scores 
})
table
```

```python
table.to_csv('/kaggle/working/table_rq6.csv', index=False)
print("table")
```

```python
plt.figure() 
plt.boxplot(scores) 
plt.title("Robustness (Cross Validation)") 
plt.savefig('/kaggle/working/figure_rq6.pdf') 
plt.show()
```
