```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestRegressor

# Load 
df = pd.read_csv("/kaggle/input/datasets/karansridhar/marketing-and-product-performance/marketing_and_product_performance.csv")

X = df.select_dtypes(include=np.number).drop(columns=['Revenue_Generated']) 
y = df['Revenue_Generated']

model = RandomForestRegressor() 
model.fit(X, y) 

importance = pd.DataFrame({ 
    "Feature": X.columns, 
    "Importance": model.feature_importances_ 
}).sort_values(by="Importance", ascending=False)
importance
```

```python
importance.to_csv('/kaggle/working/table_rq4.csv', index=False)
print("table saved")
```

```python
plt.figure() 
plt.barh(importance["Feature"][:10], importance["Importance"][:10]) 
plt.title("Top Features") 
plt.savefig('/kaggle/working/figure_rq4.pdf') 
plt.show()
```
